## Model selection based on mean square error
We first run our training script. This script will train three different regression models, namely random forest regressor, support vector machine regressor and ridge regression. Let's us look at the result of training.

```python
import pandas as pd
# MSE score for different categories usng different methods
# Methods are respectively randomForestRegressor, SVR, ridgeRegression
ele_score = pd.DataFrame({'rf':0.0356,'svr':0.1408,'rr':0.0557},
                         index=['electronic'],columns=['rf','svr','rr'])
beauty_score = pd.DataFrame({'rf':0.0301,'svr':0.1363,'rr':0.0384},
                            index = ['beauty'],columns=['rf','svr','rr'])
kindle_score=pd.DataFrame({'rf':0.056,'svr':0.1380,'rr':0.0657},
                            index=['kindle'], columns=['rf','svr','rr'])
pet_score = pd.DataFrame({'rf':0.0206,'svr':0.1669,'rr':0.0335},
                             index=['pet'], columns=['rf','svr','rr'])
music_score = pd.DataFrame({'rf':0.0374,'svr':0.1558,'rr':0.0510},
                             index=['music'], columns=['rf','svr','rr'])
office_score = pd.DataFrame({'rf':0.0339,'svr':0.1633,'rr':0.0442},
                             index=['office'] , columns=['rf','svr','rr'])

frames = [ele_score,beauty_score,kindle_score,pet_score,
          music_score,office_score]
score_df = pd.concat(frames)
score_df
```

Above table shows MSE score for each regression on model each category of review. It shows that, for each category, random forest regressor outperms other regression models. Its MSE scores range from 0.02~0.06, mostly between 0.03~0.04, which are relatively small.

Therefore, we will continue our analysis based on results of random forest regression.

## Feature importance for each category
In this section, we will show some relatively important feature based on results of random forest feature importance. We show this attribute for each category.

```python
feat_importance = main()
```

We can create a dataframe to store the importance of each feature for each category.

```python
electronic_fi_df = pd.DataFrame(feat_importance[0][0],index=['electronic'])
beauty_fi_df = pd.DataFrame(feat_importance[0][1],index=['beauty'])
kindle_fi_df=pd.DataFrame(feat_importance[0][2],index=['kindle'])
pet_fi_df = pd.DataFrame(feat_importance[0][3],index=['pet'])
music_fi_df=pd.DataFrame(feat_importance[0][4],index=['music'])
office_fi_df = pd.DataFrame(feat_importance[0][5],index=['office'])
frames_fi = [office_fi_df,electronic_fi_df,beauty_fi_df,kindle_fi_df,
            pet_fi_df,music_fi_df]
total_fi_df = pd.concat(frames_fi)
total_fi_df = total_fi_df.transpose()
total_fi_df.sort_values(by= ['office','electronic','beauty','kindle','pet','music'],ascending = False)
```

```python
rank_fi = total_fi_df.rank()
rank_fi = 33-rank_fi
rank_fi = rank_fi.sort_values(by = ['office','electronic','beauty','kindle','pet','music'],ascending = True)
rank_fi = rank_fi[['electronic','beauty','music','pet','kindle','office']]
rank_fi
```

In chart above, we find some universally important features in all categories, namely overall(the average rating of the product itself), item-rating(the score the reviewer gives to this product) and review length. They are the most three important features in all categories. 

Then, different categories might have some own different important features. Such as 'Clout' feature(a high number suggests	that the author is speaking	from the perspective of	high expertise and is confident) in office category, 'article' feature in beauty category, and 'authentic' feature (higher numbers are associated with a more honest, personal, and disclosing text) in kindle category, etc.

There are also some universally unimportant features, such as 'anger', 'heshe', 'we', etc.
